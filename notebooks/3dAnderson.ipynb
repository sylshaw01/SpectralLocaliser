```python
#import SLModels from ../src/SLmodels.py
import numpy
import os
import numpy as np
import matplotlib.pyplot as plt
import sys
sys.path.append('../src')
from SLmodels import *
```

```python
am3 = ThreeDimensionalAnderson(L=10,disorder=0.1,rho=60,kappa=0.1)

evals, evecs = am3.find_eigenvalues(am3.H,sparse=False)

```

```python
fix, ax = plt.subplots()
ax.hist(evals, bins=40, density=True, alpha=0.7, color='blue')
ax.set_xlabel("E")
ax.set_ylabel("Density of States, L=24")
ax.set_title("3D Anderson Model: Density of States L=16")
ax.grid()
plt.show()
```
